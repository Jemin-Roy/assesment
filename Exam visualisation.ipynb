```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('/content/diabetes_model.csv')
```

```python
data
```

```python
#  Pregnancies vs Glucose
data.plot(kind='scatter', x='Pregnancies', y='Glucose')
```

```python

data['Pregnancies'].plot(kind='hist',title='Pregnancies')
```

```python
plt.bar(data['Pregnancies'],data['DiabetesPedigreeFunction'])
plt.xlabel('Pregnancies')
plt.ylabel('DiabetesPedigreeFunction')
plt.title('Pregnancies & DiabetesPedigreeFunction')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.scatter(data['Pregnancies'],data['BloodPressure'])
plt.xlabel('Pregnancies')
plt.ylabel('BloodPressure')
plt.title('Pregnancies & BloodPressure')
plt.xticks(rotation=90)
plt.show()
```

```python
import seaborn as sns
```

```python
sns.pairplot(data,hue='Pregnancies')
plt.show()
```

```python
sns.histplot(data['BMI'],bins=10,kde=True)
plt.title('histogram of BMI')
plt.xlabel('BMI')
plt.ylabel('frequency')
plt.show()
```
